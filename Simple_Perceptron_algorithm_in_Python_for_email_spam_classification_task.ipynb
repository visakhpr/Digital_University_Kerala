```python
# Target Variable
y = dt['spam']
y
```

```python
from sklearn.preprocessing import MinMaxScaler
scale = MinMaxScaler()
X_transform = scale.fit_transform(X)
```

```python
X_transform = pd.DataFrame(X_transform, columns = X.columns)
X_transform
```

```python
# Splitting the Dataset into Training and Testing Dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_transform, y, test_size = 0.3)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.linear_model import Perceptron
```

```python
#Defining Model
model = Perceptron()

# Defining Model Evaluation Method
cv = RepeatedStratifiedKFold(n_splits = 20, n_repeats = 7, random_state = 42)

# Defining Grid
grid = dict()
grid['eta0'] = [0.001, 0.01,0.05, 0.1, 1.0]

# Defining Search
grid_search = GridSearchCV(model, grid, scoring = 'accuracy', cv= cv, n_jobs = 1)

# Performing the Search
results = grid_search.fit(X_train, y_train)

# Summarizing the results
print('Mean Accuracy: %.3f' % results.best_score_)
print(' Best Learning Rate: %s' %results.best_params_)
```

```python
# Summarizing the result of all Learning Rates
means = results.cv_results_['mean_test_score']
params = results.cv_results_['params']
for mean, param in zip(means, params):
  print("Mean Test Score: %.3f, Learning Rate: %r" % (mean, param))
```

We can infer from the following results that the best Mean Test Score of 0.859 was obtained when the learning rate was equal to 1.0. This suggests that model performed best when learning rate is equal to 1.0.

```python
# Defining the model
model = Perceptron(eta0=1.0)

# Defining Model Evaluation Method
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)

# Defining Grid
grid = dict()
grid['max_iter'] = [1, 10, 100, 1000, 10000]

# Defining Search
search = GridSearchCV(model, grid, scoring='accuracy', cv=cv, n_jobs=-1)

# Performing the Search
results = search.fit(X_train, y_train)

# Summarizing the Results
print('Mean Accuracy: %.3f' % results.best_score_)
print('Best Number of Maximum Iterations: %s' % results.best_params_)
```

```python
# Summarizing for all iterations
means = results.cv_results_['mean_test_score']
params = results.cv_results_['params']
for mean, param in zip(means, params):
    print("Mean Test Score: %.3f, Maximum number of Iteratations : %r" % (mean, param))
```

```python
# Creating a Perceptron Model with Best Learning Rate and Best Maximum Number of Iterations
perc_model = Perceptron(eta0 = 1.0, max_iter = 100)
perc_model.fit(X_train, y_train)
perceptron_prediction = perc_model.predict(X_test) 
```
